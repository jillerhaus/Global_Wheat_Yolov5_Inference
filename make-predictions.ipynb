```python
import glob
import re
txt_files = glob.glob('./working-yolov5/inference/output/*.txt')
jpg_files = glob.glob('../input/global-wheat-detection/test/*.jpg')
jpgfilenames = [re.compile(r'([\w\_\(\)]*)\.jpg').search(jpgfile).group(1) for jpgfile in jpg_files]
txtfilenames = [re.compile(r'([\w\_\(\)]*)\.txt').search(txtfile).group(1) for txtfile in txt_files]
```

Get image dimensions. Apparently there are currently images in the private test dataset, that are not 1024 x 1024. I am getting an error when I upload my submission. I think one possible cause of this could be my code predicting boxes outside of the pictures if their dimensions are smaller than 1024 x 1024 because yolov5 outputs the bounding boxes as relatives to width and height.

```python
import numpy as np
import pandas as pd
import cv2
img_dimensions = [cv2.imread(image).shape for image in jpg_files]
dims = pd.DataFrame(img_dimensions, columns = ['Width', 'Height', '3'])
dims.drop(columns = ['3'], inplace = True)
dims.shape
filenames = pd.DataFrame(jpgfilenames, columns = ['filename'])
pic_atrs = pd.concat([filenames, dims], axis = 1)
num_pics = pic_atrs.shape[0]

num_pics
```

```python
submission = pd.DataFrame(columns = ['image_id', 'PredictionString'])
i = 0
for i in range(num_pics):
    filename = pic_atrs.iloc[i].filename
    width = pic_atrs.iloc[i].Width
    height = pic_atrs.iloc[i].Height
    if filename in txtfilenames:
        dataset = pd.read_csv("working-yolov5/inference/output/" + filename + '.txt', header = None, delimiter = ' ')
        dataset.drop(columns = 5, inplace = True)
        dataset.columns = ['conf', 'x_mid', 'y_mid', 'box_w', 'box_h']
        dataset['width'] = width
        dataset['height'] = height
        dataset['px_min'] = dataset.x_mid*width - 0.5*dataset.box_w*width
        dataset['py_min'] = dataset.y_mid*height - 0.5*dataset.box_h*height
        dataset['px_box'] = dataset.box_w * width
        dataset['py_box'] = dataset.box_h * height
        dataset[['px_min', 'py_min', 'px_box', 'py_box']] = dataset[['px_min', 'py_min', 'px_box', 'py_box']].round().astype('int')
        dataset.loc[dataset.width < dataset.px_min,'px_min'] = dataset.width -1
        dataset.loc[dataset.height < dataset.py_min,'py_min'] = dataset.height -1
        dataset.loc[1 > dataset.px_box,'px_box'] = 1
        dataset.loc[1 > dataset.py_box,'py_box'] = 1
        dataset.loc[0 > dataset.px_min,'px_min'] = 0
        dataset.loc[0 > dataset.py_min,'py_min'] = 0
        dataset.loc[dataset.px_min + dataset.px_box > dataset.width, 'px_box'] = dataset.width - dataset.px_min - 1
        dataset.loc[dataset.py_min + dataset.py_box > dataset.height, 'py_box'] = dataset.height - dataset.py_min - 1
        submission_cols = ['conf', 'px_min', 'py_min', 'px_box', 'py_box']
        shape = dataset[submission_cols].shape
        row = dataset[submission_cols].values.reshape(1,shape[0]*shape[1])
        str_array = np.array2string(row[0], formatter = {'float_kind': lambda x: '%.5f' % x}, max_line_width = 9e3)
        str_array = re.compile(r'\[([\d\s\.]+)\]').search(str_array).group(1)
#         str_array = re.compile(r'\[(.+)\]').search(str_array).group(1)
        vals = pd.DataFrame([[filename, str_array]], columns =['image_id', 'PredictionString'])
        
    else:
        vals = pd.DataFrame([[filename, '']], columns =['image_id', 'PredictionString'])
    
    submission = pd.concat([submission,vals], axis = 0)
    submission.PredictionString = submission.PredictionString.str.replace(r'(\.0+)', '').str.replace('\s+',' ')
submission.to_csv('submission.csv', index = False, sep = ',')
```
